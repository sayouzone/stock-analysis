```python
from io import StringIO
import pandas as pd

import requests
from requests import HTTPError
from bs4 import BeautifulSoup
```

```python
stock = '005930'
url = f"https://comp.fnguide.com/SVO2/ASP/SVD_Finance.asp?pGB=1&gicode=A{stock}&cID=&MenuYn=Y&ReportGB=&NewMenuID=103&stkGb=701"
```

```python
response = requests.get(url)
response.raise_for_status()

soup = BeautifulSoup(response.text, 'html.parser')

tables = soup.find_all('table')

dfs = []
if tables != []:
    for table in tables:
        # HTML 객체를 문자열로 변환 후 데이터프레임으로 변환(StringIO는 필수는 아니지만 터미널 로그 정리 목적)
        df = pd.read_html(StringIO(str(table)))
        dfs.append(df)
```
